**Carregamento dos Dados**

Abaixo vamos carregar os dados

```python
import pandas as pd

pib_municipios = pd.read_excel('PIB dos Municípios - base de dados 2010-2021.xlsx')
```

Abaixo, é mostrado as 5 primeiras observações desse conjunto de dados

```python
pib_municipios.head()
```

Optamos por fazer a análise descritiva dos dados apenas do nosso estado - Goiás, abaixo selecionamos apenas as tuplas em 
que a Sigla da Unidade da Federação é GO.

```python
pib_municipios_goias = pib_municipios[pib_municipios['Sigla da Unidade da Federação'] == 'GO']
```

Abaixo, a visualização de algumas tuplas desse conjunto de dados.

```python
pib_municipios_goias
```

```python
pib_municipios_goias.head()
```

Abaixo os dados da nossa cidade, Goiânia

```python
# Sua condição gera uma série de True/False
filtro_goiania = pib_municipios_goias["Nome do Município"] == 'Goiânia'

# Agora use esse filtro para selecionar as linhas do DataFrame original
linha_goiania = pib_municipios_goias[filtro_goiania]

# Imprima o resultado
linha_goiania
```

Agora explorando os dados podemos ver muitas colunas e para uma análise, damos maior atenção a algumas dessas colunas

```python
print(pib_municipios_goias.columns)
```

Vamos selecionar as colunas para a análise

```python
colunas_selecionadas = [
    'Ano', 'Nome da Unidade da Federação', 'Nome do Município',
    'Nome da Grande Região', 'Hierarquia Urbana',
    'Atividade com maior valor adicionado bruto',
    'Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)',
    'Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)',
    'Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)',
    'Valor adicionado bruto da Administração, defesa, educação e saúde públicas e seguridade social, \na preços correntes\n(R$ 1.000)',
    'Produto Interno Bruto, \na preços correntes\n(R$ 1.000)',
    'Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)'
]

novos_nomes_colunas = {'Ano': 'Ano',
                       'Nome da Unidade da Federação': 'UF',
                       'Nome do Município': 'Município',
                       'Nome da Grande Região': 'Grande Região',
                       'Hierarquia Urbana': 'Hierarquia',
                       'Atividade com maior valor adicionado bruto': 'Atividade Principal',
                       'Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)': 'VAB Agropecuária (R$ 1.000)',
                       'Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)': 'VAB Indústria (R$ 1.000)',
                       'Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)': 'VAB Serviços (R$ 1.000)',
                       'Valor adicionado bruto da Administração, defesa, educação e saúde públicas e seguridade social, \na preços correntes\n(R$ 1.000)': 'VAB Administração Pública (R$ 1.000)',
                       'Produto Interno Bruto, \na preços correntes\n(R$ 1.000)': 'PIB (R$ 1.000)',
                       'Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)': 'PIB per capita (R$)'}

pib_municipios_goias_analise = pib_municipios_goias[colunas_selecionadas]
pib_municipios_goias_analise.rename(columns=novos_nomes_colunas, inplace=True)
```
