**Carregamento dos Dados**

Abaixo vamos carregar os dados

```python
import pandas as pd

pib_municipios = pd.read_excel('PIB dos Municípios - base de dados 2010-2021.xlsx')
```

Abaixo, é mostrado as 5 primeiras observações desse conjunto de dados

```python
pib_municipios.head()
```

Optamos por fazer a análise descritiva dos dados apenas do nosso estado - Goiás, abaixo selecionamos apenas as tuplas em 
que a Sigla da Unidade da Federação é GO.

```python
pib_municipios_goias = pib_municipios[pib_municipios['Sigla da Unidade da Federação'] == 'GO']
```

Abaixo, a visualização de algumas tuplas desse conjunto de dados.

```python
pib_municipios_goias
```

```python
pib_municipios_goias.head()
```

Abaixo os dados da nossa cidade, Goiânia

```python
# Sua condição gera uma série de True/False
filtro_goiania = pib_municipios_goias["Nome do Município"] == 'Goiânia'

# Agora use esse filtro para selecionar as linhas do DataFrame original
linha_goiania = pib_municipios_goias[filtro_goiania]

# Imprima o resultado
linha_goiania
```
